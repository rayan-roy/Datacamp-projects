## Challenge 1: Find the six month rolling average (a.k.a. simple moving average) for each date and browser in the dataset

To find rolling average, we first pick a month, and then calculate the average for 6 more dates before that

```python
rolling_six = df.rolling(window=6).mean()
rolling_six.head(10)
```

Notice how the first 5 dates are NaN, its because for rolling average you need all 6 month, and 1st 6 months is on month 6, and then it becomes month 2 to 7 as another 6 months

```python
rolling_six.plot(title="6 Month Rolling Avg") # We can see the plot is much smoother, than earlier one
df.plot()
```

## Challenge 2: Create a DataFrame called pct_change_quarterly with the percentage change from the previous quarter for each date and browse

```python
# Since chrome launched in late 2008, we include dates after that

# We put in 3 because we are skipping 3 months, and multiply by 100 to get as percentage
pct_change_quarterly = df.pct_change(3)*100
pct_change_quarterly.head()

# The way they calculate, April has 20 and Jan has 12, so (20-12)/12 = 66%
```

```python
# Remember how we had kept col_index as month, which means we can directly use 2009
pct_change_quarterly = pct_change_quarterly.loc['2009':]
pct_change_quarterly.plot()
# find what loc function does
```

```python
# Too hard to read
pct_change_quarterly.plot(subplots=True, figsize = (12,8))
```

We can see in 2016, Firefox and Chrome had a big dip, maybe something happened?
As for explorer, it never grew, and in fact lost customers in 2019 and 2020.


Although from the df.plot we see that chrome overtook explorer in 2012, the individual graphs don't show that Chrome grew that much. Further exploration is required. We will look at Chrome's trend in 2009, 2012, 2015 and 2018

```python
chrome_trends = pd.DataFrame()
for year in ['2009', '2012', '2015', '2018']:
    chrome_trends_per_year = df.loc[year, ['Google Chrome']].reset_index(drop = True)
    # we will take 2009 year and then specifically look for Chrome
    chrome_trends_per_year.rename(columns = {'Google Chrome': year}, inplace = True)
    chrome_trends = pd.concat([chrome_trends, chrome_trends_per_year], axis = 1)
    # axis = 1 puts the trends_per_year beside the trends
    print(chrome_trends)
    
    # My annotations: 
    # We first take the year 2009, and then later add 2012 b/c of list of [2009,2012]
    # The loc first takes a row, and then a column (Chrome)
```

```python
chrome_trends.plot(title="Chrome Search Performance in 2009, 2012, 2015 & 2018")
```

As we can see in 2009, chrome was growing rapidly each month, while in 2015 it was decling every month. 2018 and 2012 seem very flat
